```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('WeatherTrainingData.csv')
data.drop(labels=['Location','Evaporation', 'WindDir9am', 'WindDir3pm' ,'WindGustDir','Sunshine' ,'row ID' ], inplace=True, axis=1)
data["Cloud3pm"].fillna(0, inplace = True)
data["Cloud9am"].fillna(0, inplace = True)
data.replace({'RainToday': {"No": 0, 'Yes': 1}},inplace=True)
data.dropna(inplace=True)
data.reset_index(drop=True, inplace=True)
data
```

```python

values=data.select_dtypes(exclude="object")
columnNames=values.columns.tolist()
columnNames=columnNames[:-2]
for column in columnNames:
        x=data.loc[:,column]
        min1=min(x)
        max1=max(x)
        for row in range (len(data)):
            xprim=(data.at[row,column]-min1)/(max1-min1)
            data.at[row,column]=xprim
data
```

```python
data.to_csv('Weather-fixed.csv')
```
