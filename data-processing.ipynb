```python
import pandas as pd
import numpy as np
import seaborn as sns
import math 
import random as rn
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
def split(x, k):  
    splitPoint = int(len(x) * k)
    return x.iloc[:splitPoint], x.iloc[splitPoint:]

def getDistances(x, sample, columnNames, function, power):
    lis = []
    for i in range(len(x)):
        y=0
        for column in columnNames:
            y+=(function(x.loc[i,column], sample[column], power))
        lis.append(y)
    return lis


def manhattan(first, second, n):
    return abs(first - second)
        

def euclides(first, second, n):
    difference = math.pow(first, n) - math.pow(second, n)
    return difference ** 1/n

def sort(x, lis):
        x['distance'] = lis
        return x.sort_values(by=['distance'])

def shuffle(x):
    for i in range(len(x) - 1, 0, -1):
        j = rn.randint(0, i - 1)
        x.iloc[i], x.iloc[j] = x.iloc[j], x.iloc[i]

def knn(recs,k):
    dict={1:0,0:0}
    knn=recs.iloc[:k]['RainTomorrow']
    for i in knn:
        if i:
            dict[1]+=1
        else:
            dict[0]+=1
    return (max(dict, key=dict.get))

def knnRain(recs,k):
    
    
    rain=recs.iloc[:k]['Rainfall']
    mean = rain.mean()
    sigm = rain.std()
  
    return mean,sigm

def knnTemp(recs,k):  
    rain=recs.iloc[:k]['AvgTemp']
    mean = rain.mean()
    sigm = rain.std()
  
    return mean,sigm

def avgTemp(x):
    lis = []
    for i in range(len(x)):
        lis.append(x.iloc[i]['Temp3pm']+x.iloc[i]['Temp9am']*0.5)
    x['AvgTemp'] = lis
```

```python
x = data.drop('Rainfall', axis=1).values
y = [int(value) for value in data['Rainfall']]

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42 )
```

```python
k_values = [i for i in range(1,31)]
accuracies = []
for value in k_values:
    knn = KNeighborsClassifier(n_neighbors=value)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    accuracies.append(accuracy_score(y_test, y_pred))

accuracies
```
