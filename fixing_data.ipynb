```python
import pandas as pd
import numpy as np
import seaborn as sns
import math 
import random as rn
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('WeatherTrainingData.csv')
data.drop(labels=['Location','Evaporation', 'WindDir9am', 'WindDir3pm' ,'WindGustDir','Sunshine','row ID', 'RainTomorrow' ], inplace=True, axis=1)
data["Cloud3pm"].fillna(0, inplace = True)
data["Cloud9am"].fillna(0, inplace = True)
data.replace({'RainToday': {"No": 0, 'Yes': 1}},inplace=True)
data.dropna(inplace=True)
data.reset_index(drop=True, inplace=True)
data = data[:len(data)//2]
data
```

```python
avgtemp = [0]*len(data)
avgwind = [0]*len(data)
avghumidity = [0]*len(data)
avgpressure = [0]*len(data)
avgclouds = [0]*len(data)

len(avgtemp)
for i in range (len(data)):
    avgtemp[i] = (data.iloc[i]['Temp3pm']+data.iloc[i]['Temp9am'])*0.5
    avgwind[i] = (data.iloc[i]['WindSpeed9am']+data.iloc[i]['WindSpeed3pm'])*0.5
    avghumidity[i]=(data.iloc[i]['Humidity9am']+data.iloc[i]['Humidity3pm'])*0.5
    avgpressure[i]=(data.iloc[i]['Pressure9am']+data.iloc[i]['Pressure3pm'])*0.5
    avgclouds[i]=(data.iloc[i]['Cloud9am']+data.iloc[i]['Cloud3pm'])*0.5
data['AvgTemp'] = avgtemp
data['AvgWind'] = avgwind
data['AvgHumidity'] = avghumidity
data['AvgPressure'] = avgpressure
data['AvgClouds'] = avgclouds
data.drop(labels=['Temp3pm','Temp9am','WindSpeed9am', 'WindSpeed3pm', 'Humidity9am',
                   'Humidity3pm',  'Pressure9am',  'Pressure3pm', 'Cloud9am', 'Cloud3pm'], inplace=True, axis=1)
data
```

```python
cols_to_norm = ['MinTemp', 'MaxTemp', 'WindGustSpeed', 'AvgTemp', 'AvgWind', 'AvgHumidity', 'AvgPressure', 'AvgClouds']
data[cols_to_norm] = data[cols_to_norm].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
data
```

```python
data.to_csv('Normalized_data.csv')
```
